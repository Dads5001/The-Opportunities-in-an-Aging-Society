```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib as mpl


mpl.rcParams['font.family'] = 'Arial'
labels = ['Adequacy', 'Sustainability', 'Integrity']
sizes = [40, 35, 25]
pastel_colors = ['#a1c9f4', '#ffb482', '#8de5a1']


fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, counterclock=False, colors=pastel_colors, textprops={'fontsize': 12})
ax.set_title('The Mercer CFA Institute Global Pension Index', fontsize=14)

plt.savefig('index.png', dpi=2000)
```

```python
df = pd.read_csv("index.csv")
df = df.pivot(index='country', columns='year', values='index_value')

sns.set(font_scale=0.4, font='Arial')

plt.figure(dpi=200)
sns.heatmap(data=df, annot=True, fmt='.1f', cmap='rocket')

plt.savefig('heatmap.png', dpi=2000)
```
