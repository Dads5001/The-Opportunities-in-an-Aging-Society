```python
#compare Thai and Japan
import sys
import numpy as np
import pandas as pd

import matplotlib as mpl
%matplotlib inline
import matplotlib.pyplot as plt


import seaborn as sns
sns.set()  

print( f"Python {sys.version}" )
print( f"Pandas {pd.__version__}" )
print( f"NumPy {np.__version__}" )
print( f"Matplotlib {mpl.__version__}" )
print( f"Seaborn {sns.__version__}" )
```

```python
data = {'year': [2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022],
        'population': [62952762, 63338438, 64785909, 65124716, 65729098, 65931550, 66188503, 66413979, 66558935, 66186727, 66171439, 66090475],
        'elderly_population': [7811450, 8170909, 9057761, 9446027, 9814833, 10159630, 10599080, 11058037, 11547355, 11789098, 12071837, 12698329],
        'elderly_rate': [12.4, 12.9, 14.0, 14.5, 14.9, 15.4, 16.0, 16.7, 17.3, 17.8, 18.2, 19.2]}

df = pd.DataFrame(data)

print(df)
```

```python
sns.set_style('ticks')



plt.figure(figsize=(12, 8))
sns.set_style('whitegrid')


sns.lineplot(data=df, x='year', y='elderly_population', color='purple', linestyle='--')


plt.xlabel('Year')
plt.ylabel('Elderly Population')
plt.title('Trend of Elderly Population in Thailand')


plt.show()
```
