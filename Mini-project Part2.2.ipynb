```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('index.csv')
df = pd.DataFrame(data)

df = df[(df['year'] == 2022) & df['country'].isin(['Thailand', 'Iceland'])]

grouped = df.groupby('country').sum()

fig, axs = plt.subplots(1, 3, figsize=(12, 4))

colors = ['#87CEFA', '#FFA07A']

axs[0].pie(grouped['adequacy'], labels=grouped.index, autopct='%1.1f%%', colors=colors)
axs[0].set_title('Adequacy')
axs[1].pie(grouped['sustainability'], labels=grouped.index, autopct='%1.1f%%', colors=colors)
axs[1].set_title('Sustainability')
axs[2].pie(grouped['integrity'], labels=grouped.index, autopct='%1.1f%%', colors=colors)
axs[2].set_title('Integrity')

plt.savefig('ratioIcelandandThailand.png', dpi=2000)
```

```python
df = pd.read_csv('index.csv')

countries = ['Thailand']
years = [2019, 2020, 2021, 2022]
df_th = df[(df['country'].isin(countries)) & (df['year'].isin(years))]

df_avg_ade = df.groupby('year')['adequacy'].mean().reset_index()
df_avg_sus = df.groupby('year')['sustainability'].mean().reset_index()
df_avg_int = df.groupby('year')['integrity'].mean().reset_index()

sns.set_style('whitegrid')
fig, axs = plt.subplots(1, 3, figsize=(16, 5))

sns.lineplot(x='year', y='adequacy', data=df_th, color='blue', label='Thailand', ax=axs[0])
sns.lineplot(x='year', y='adequacy', data=df_avg_ade, color='black', label='Average', ax=axs[0])
axs[0].set_title('Adequacy')

sns.lineplot(x='year', y='sustainability', data=df_th, color='green', label='Thailand', ax=axs[1])
sns.lineplot(x='year', y='sustainability', data=df_avg_sus, color='black', label='Average', ax=axs[1])
axs[1].set_title('Sustainability')

sns.lineplot(x='year', y='integrity', data=df_th, color='orange', label='Thailand', ax=axs[2])
sns.lineplot(x='year', y='integrity', data=df_avg_int, color='black', label='Average', ax=axs[2])
axs[2].set_title('Integrity')

plt.suptitle('Index Value for Thailand from 2019 to 2022', fontsize=14)
fig.text(0.07, 0.5, 'Index Value', va='center', rotation='vertical')


plt.savefig('graph.png', dpi=2000)
```
